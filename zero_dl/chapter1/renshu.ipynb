numpyの行列演算はややわかりにくい

```python
import numpy as np

A=np.array([[1,2],[3,4]])
B=np.array([10,20])
A*B
#broadcast




A=np.array([[1,2],[3,4]])
C=A
# Broadcasting
print(A*C)

# Dot product
print(A@C)

```

cosなども追加するとより分かりやすい
gnuplotを触る要領で行ける。

```python
import numpy as np
import matplotlib.pyplot as plt

#creating data (numpy)
x=np.arange(0,6,0.1)

y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1,label="sin")
plt.plot(x,y2,linestyle="--",label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
